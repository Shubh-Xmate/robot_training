```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import losses as lo_ss
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense, Dropout
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
import seaborn as sns
from sklearn import model_selection
```

```python
import os
github_raw_link="https://raw.githubusercontent.com/Aryansh085/LAP/main/858417d1-7d54-4115-a01b-fdda5e03ada3_testing_combined_rows4035_disc_1_0p9_MULTIPLY_preproces_155cols.csv"
df=pd.read_csv((github_raw_link),header = None)

df
```

```python
y=np.array(df[0])
```

```python
y.shape
```

```python
x=np.array([df[i] for i in range(5,155)])
print(x.T.shape)
```

```python
x_train,x_test,y_train,y_test=model_selection.train_test_split(x.T,y)
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
slr=StandardScaler()
slr.fit(x_train)
x_train = slr.transform(x_train)
x_test = slr.transform(x_test)
print(y_train.shape)
y_train=tf.keras.utils.to_categorical(y_train)
y_test=tf.keras.utils.to_categorical(y_test)
```

```python
model=Sequential()
model.add(Dense(128,input_shape=(150,)))
model.add(Dropout(0.2))
model.add(Dense(64,activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(3))
```

```python
print(set(y))
print(y_train.shape)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
model.compile(loss='mean_squared_error', optimizer='adam')
```

```python
model.fit(x_train, y_train, epochs=4, batch_size = 1)
```

```python
model.evaluate(x_train,y_train)
```

```python
model.evaluate(x_test,y_test)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
y_pre=model.predict(x_test)
```

```python
print(y_pre.shape)
```

```python
y_pre_ans=np.array([np.argmax(i)  for i in y_pre])
```

```python
print(set(y_pre_ans))
```

```python
y_test_ans=np.array([np.argmax(i)  for i in y_test])
```

```python
print(confusion_matrix(y_test_ans,y_pre_ans))
```
